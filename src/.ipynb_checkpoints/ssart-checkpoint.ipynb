```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
# Import image
img = cv.imread('./imgs/cat.jpg')
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
# plt.imshow(img)
```

```python
# thresh1 = img_gray
thresh_type = [cv.THRESH_BINARY, cv.THRESH_TRUNC]
ret,thresh1 = cv.threshold(img_gray,135,255,thresh_type[0])
# plt.imshow(thresh1)
```

```python
# Take sample size as input
sample_size = 10000
```

```python
input_shape = thresh1.shape
print(input_shape)
sample_points = np.array([np.random.randint(0,input_shape[0], sample_size), \
                          np.random.randint(0,input_shape[1], sample_size)]).transpose()
```

```python
# Now based on the color value choose a color for plotting
# Sample the color values for sampled pixels
color_values = []
for sample_point in sample_points:
    color_values.append(thresh1[sample_point[0], sample_point[1]])
```

```python
# Filtering unique values
res_colors = [];
[res_colors.append(x) for x in color_values if x not in res_colors];
```

```python
# Choose the number color bands same as 
# color_index = np.random.choice(range(256), size=len(res_colors))
```

```python
# Since creating an image is very high-resolution --> Rather just plot
# Seperate channels for plotting
indicies = []

for sample_point in sample_points:
    for i in range(len(res_colors)):
        if thresh1[sample_point[0], sample_point[1]] == res_colors[i]:
            indicies.append(np.append(sample_point, i))
indicies = np.array(indicies)
```

```python
fig = go.Figure()
marker_size = 5
for i in range(len(res_colors)):
    
    index_values = np.where(indicies == i)
    selected_samples = sample_points[index_values[0]]
    fig.add_trace(go.Scatter(x=selected_samples[:, 1], y=selected_samples[:, 0], \
                             mode='markers',\
                         marker=dict(size=marker_size)))


fig.update_yaxes(range = [input_shape[0],0])
fig.update_xaxes(range = [0, input_shape[1]])
fig.show()
```
